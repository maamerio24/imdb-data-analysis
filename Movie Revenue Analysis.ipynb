## Exploratory Data Analysis

> Use Matplotlib to display your data analysis

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python
#found this nifty trick online to combine setting the column = str and then exploding all in one step.
df_production = df_production.assign(production_companies=df.production_companies.str.split("|")).explode('production_companies').reset_index(drop=True)
```

```python
busy_bees = df_production.groupby(['production_companies']).count()[['id']]
best_busy_bees = busy_bees.sort_values(by = ['id'],ascending=False).head(5)

best_busy_bees.plot.barh(xlabel='Production Companies', ylabel='Number of Movies Released', grid=True, title='Production Companies with Most Releases', figsize=(15,10))
plt.legend(["# of releases"]);
```

### What 5 movie genres grossed the highest all-time?

```python
df_genre = df_genre.assign(genres=df.genres.str.split("|")).explode('genres').reset_index(drop=True)
```

```python
highest_grossed = df_genre.groupby(['genres']).sum()[['revenue']]
money_makers = highest_grossed.sort_values(by = ['revenue'], ascending=False).head(5)


money_makers.plot.bar(xlabel='Genres', ylabel='($) Grossed', grid=True, title='Hghest Grossing Genres', figsize=(15,10))
plt.legend(["# of releases"]);
```

### Who are the top 5 grossing directors?

```python
goat_directors = df.groupby(['director']).sum()[['revenue']]
directors_pie = goat_directors.sort_values(by = ['revenue'], ascending=False).head(5)

directors_pie.plot.pie(subplots=True, figsize=(15, 14));
```

### Compare the revenue of the highest grossing movies of all time.

```python
goat_movies = df.groupby(['original_title']).sum()[['revenue']]
most_pop_movies = goat_movies.sort_values(by = ['revenue'], ascending=False).head(10)
most_pop_movies.plot.pie(subplots=True, figsize=(20, 20));
```
